```r
# 1
time_2400_to_minutes <- function(t) {
    hours <- t %/% 100
    minutes <- t %% 100
    return(hours * 60 + minutes)
}
```

```r
flights %>%
select(dep_time, sched_dep_time) %>%
mutate(dep_minutes = time_2400_to_minutes(dep_time),
      sched_dep_minutes = time_2400_to_minutes(sched_dep_time))
```

```r
names(flights)
```

```r
# 2
flights %>%
select(air_time, arr_time, dep_time) %>%
mutate(dep_to_arr_time = time_2400_to_minutes(arr_time) - time_2400_to_minutes(dep_time)) %>%
mutate(time_descrepancy = air_time - dep_to_arr_time)
```

```r
# 3
calculate_delay <- function(sched_dep_time, dep_time) {
    sched_dep_time <- time_2400_to_minutes(sched_dep_time)
    dep_time <- time_2400_to_minutes(dep_time)
    delay <- ifelse(dep_time - sched_dep_time < -60, 
                    dep_time + 1440 - sched_dep_time,
                    dep_time - sched_dep_time)
    return(delay)
}
```

```r
df3 <- flights %>%
select(sched_dep_time, dep_time, dep_delay) %>%
mutate(delay_calculated = calculate_delay(sched_dep_time, dep_time)) %>%
mutate(time_discrepancy = dep_delay - delay_calculated) %>%
filter(!is.na(time_discrepancy))
```

```r
filter(df3, time_discrepancy != 0)
```

```r
# 4
flights %>%
mutate(total_delay = dep_delay + arr_delay) %>%
mutate(delay_rank = min_rank(desc(total_delay))) %>%
arrange(delay_rank)
```

# Useful summary functions

```r
not_cancelled <- flights %>%
filter(!is.na(dep_time), !is.na(arr_time))
```

```r
df1 <- not_cancelled %>%
group_by(year, month, day) %>%
mutate(r = min_rank(desc(dep_time))) %>%
select(year, month, day, dep_time, r)

filter(df1, r == 1)
```

```r
daily <- group_by(flights, year, month, day)
per_day <- summarize(daily, flights = n())
class(per_day)
```

```r
# 2
df1 <- not_cancelled %>%
count(dest)
head(df1)
```

```r
df2 <- not_cancelled %>%
group_by(dest) %>%
summarize(n = n())
head(df2)
```

```r
df3 <- not_cancelled %>%
count(tailnum, wt = distance)
head(df3)
```

```r
df4 <- not_cancelled %>%
group_by(tailnum) %>%
summarize(n = sum(distance))
head(df4)
```

```r
any_na <- not_cancelled[rowSums(is.na(not_cancelled)) != 0,]
head(any_na)
```
